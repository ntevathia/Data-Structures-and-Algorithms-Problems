```python
class Node:
    def __init__(self, value, next = None):
        self.value = value
        self.next = next

class MyLinkedList(object):

    def __init__(self):
        """
        Initialize your data structure here.
        """
        self.root = None
        
    def get(self, index):
        """
        Get the value of the index-th node in the linked list. If the index is invalid, return -1.
        :type index: int
        :rtype: int
        """
        
        cur = self.root
        count = 0
        
        if self.root is None and index>0:
            print("empty list and position not found")
            return -1
        
        while cur is not None and index!=count:
            cur = cur.next
            count+=1
        if cur is None:
            print("position not found")
            return -1
        if index == count:
            return cur.value
        
        
        

    def addAtHead(self, val):
        """
        Add a node of value val before the first element of the linked list. After the insertion, the new node will be the first node of the linked list.
        :type val: int
        :rtype: None
        """
        node = Node(val) 
        # if no node exists
        if self.root is None:
            self.root = node
        else:
            # if list exists already
            node.next = self.root
            self.root = node
            

    def addAtTail(self, val):
        """
        Append a node of value val to the last element of the linked list.
        :type val: int
        :rtype: None
        """
        node = Node(val) 
        # if list does not exists
        if self.root is None:
            self.root = node
        else:
            # otherwise move till the end and append the node
            cur = self.root
            while cur.next is not None:
                cur = cur.next
            cur.next = node
   

    def addAtIndex(self, index, val):
        """
        Add a node of value val before the index-th node in the linked list. If index equals to the length of linked list, the node will be appended to the end of linked list. If index is greater than the length, the node will not be inserted.
        :type index: int
        :type val: int
        :rtype: None
        """
        node = Node(val) 
        cur = self.root
        count = 0
        
        if self.root is None and index>0:
            print("empty list and position not found")
            return
        if index==0:
            return self.addAtHead(val)
        
        
        while cur is not None and index!=count:
            prev = cur
            cur = cur.next
            count+=1
            
        
        #print("count ", count)
        if (cur is None and index==count) or (index == count):
            print("inserting")
            node.next = cur
            prev.next = node
            return 
        if cur is None:
            print("position not found")
            return
        
        

    def deleteAtIndex(self, index):
        """
        Delete the index-th node in the linked list, if the index is valid.
        :type index: int
        :rtype: None
        """
        cur = self.root
        count = 0
        
        # if list is empty
        if self.root is None:
            print("empty list")
            return
        
        # only one elemnt present in the list
        if self.root.next is None:
            if index == count:
                self.root = None
                return
            else:
                print("position not found")
                return
        
        # handles 1st deletion in the list
        if index == 0:
            cur = cur.next
            self.root = cur
            return
        
        while cur is not None and index!=count:
            prev = cur
            cur = cur.next
            count = count+1
        
        # when index is not in the list
        if cur is None and index>=count:
            print("index not found")
            return
            
        # regular case
        if index == count and cur is not None:
            prev.next = cur.next
            return
        
        
```
