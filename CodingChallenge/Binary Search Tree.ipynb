BSTs are of interest because they have operations which are favourably fast: insertion, look up, and deletion can all be done in O(log n) time.

```python
from lolviz import treeviz
```

```python
class Node:
    def __init__(self, value, left=None, right=None):
        self.value=value
        self.left=left
        self.right=right
```

### Insertion

```python
def insertBST(root, value):
    node = Node(value)
    if root==None:
        root = node
    elif value<root.value:
        if root.left==None:
            root.left=node
        else:
            insertBST(root.left, value)
    elif value>=root.value:
        if root.right==None:
            root.right=node
        else:
            insertBST(root.right, value)
    return root
```

```python
root = insertBST(root, 10)
```

```python
root = insertBST(root, 20)
root = insertBST(root, 30)
root = insertBST(root, 5)
root = insertBST(root, 6)
root = insertBST(root, 11)
```

### Searching

```python
def searchBST(root, value):
    if root==None:
        return False
    if root.value==value:
        return True
    elif value<root.value:
        return searchBST(root.left, value)
    elif value>root.value:
        return searchBST(root.right, value)
    
```

```python
searchBST(root, 99)
```

### Find parent node

```python
def findParent(root, value):
    if root==None:
        print("Empty tree!")
        return None
    elif root.value==value:
        return False
    elif value<root.value:
        if root.left==None:
            return False
        elif root.left.value==value:
            return root
        else:
            return findParent(root.left, value)
    else:
        if root.right==None:
            return False
        elif root.right.value==value:
            return root
        else:
            return findParent(root.right, value)
    
```

### Find the node

```python
def findNode(root, value):
    if root==None:
        return None
    if root.value==value:
        return root
    elif value<root.value:
        return findNode(root.left, value)
    elif value>root.value:
        return findNode(root.right, value)
    
```

```python
node = findNode(root, 30)
print(node.value)
```

```python
node = findNode(root, 99)
print(node)
```

### Deletion

```python
def deleteBST(root, value):
    if root==None: 
        return None

    if value < root.value: 
        root.left = deleteBST(root.left, value) 
    elif value > root.value: 
        root.right = deleteBST(root.right, value) 
    else: # if value==root.value
        # when either of left or right subtree is missing
        if root.left==None : 
            new_root = root.right    # make right tree root as root
            root = None 
            return new_root    
        elif root.right is None : 
            new_root = root.left  
            root = None
            return new_root
  
        # when both subtrees are present 
        cur = root.right
        while(cur.left!=None): # find the smallest value in right subtree
            cur = cur.left 
  
        root.value = cur.value # make root as current and delete current
        root.right = deleteBST(root.right , cur.value) 
  
    return root  
```

### find smallest and larget value in the tree

```python
def findSmall(root):
    cur = root
    while cur.left!=None:
        cur=cur.left
    return cur.value
```

```python
def findLarge(root):
    cur = root
    while cur.right!=None:
        cur=cur.right
    return cur.value
```

```python
findSmall(root)
```

```python
findLarge(root)
```

### Tree Traversals

```python
def inOrder(root):
    if root==None:
        return None
    inOrder(root.left)
    print(root.value)
    inOrder(root.right)
```

```python
def preOrder(root):
    if root==None:
        return None
    print(root.value)
    preOrder(root.left)
    preOrder(root.right)
```

```python
def postOrder(root):
    if root==None:
        return None
    postOrder(root.left)
    postOrder(root.right)
    print(root.value)
```

```python
treeviz(root)
```

```python
inOrder(root)
```

```python
preOrder(root)
```

```python
postOrder(root)
```

### BreadthFirst
- using queue

```python
from collections import deque
```

```python
def breadthFirstTraversal(root):
    q = deque()
    while root!=None:
        print(root.value)
        if root.left!=None:
            q.append(root.left)
        if root.right!=None:
            q.append(root.right)
        
        if q!=deque([]):
            root = q.popleft()
        else:
            root=None
```

```python
breadthFirstTraversal(root)
```
