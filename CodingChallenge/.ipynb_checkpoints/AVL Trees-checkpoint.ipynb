"the first self balancing binary search tree data structure, calling it AVL Tree."

An AVL tree is a binary search tree with a self-balancing condition stating that the difference between the height of the left and right subtrees cannot be no more than one

The BST represents the worst case scenario in which the running time of all common operations such as search, insertion and deletion are O(n). By applying a balance condition we ensure that the worst case running time of each common operation is O(log n). The height of an AVL tree with n nodes is O(log n) regardless of the order in which values are inserted.

### Tree rotation to maintain balanced BST

```python
def LeftRotation(root):
    if root.right!=None:
        rightnode = root.right
        root.right = rightnode.left
        rightnode.left = root
        return rightnode
    return root   
```

```python
def RightRotation(root):
    if root.left!=None:
        leftnode = root.left
        root.left = leftnode.right
        leftnode.right = root
        return leftnode
    return root   
```

```python
class Node:
    def __init__(self, value, left=None, right=None):
        self.value=value
        self.left=left
        self.right=right
```

```python
head = Node(14)
a = Node(8)
b = Node(2)
c = Node(11)
d = Node(24)
head.left=a
head.right = d
a.left=b
a.right=c
```

```python
## O(1)
```

```python
def height(root):
    if root is None:
        return 0
    else:
        lt=height(root.left)
        rt=height(root.right)
        if lt>rt:
            return lt+1
        else:
            return rt+1       
```

```python
height(head)
```

```python
def checkBalance(root):
    if height(root.left)-height(root.right)>1:
        root = RightRotation(root)
    elif height(root.right)-height(root.left)>1:
        root = LeftRotation(root)
    return root
```

```python
aa = Node(1)
bb = Node(6)
b.left=aa
b.right=bb
```

```python
lolviz.treeviz(head)
```

```python
root = checkBalance(head)
```

```python
lolviz.treeviz(root)
```

```python
root = checkBalance(root)
```

### Insert
- Add values with BST property
- Check balanced height for AVL

```python
def insertAVL(root, value):
    if root is None:
        node = Node(value)
        root = node
    else:
        node = Node(value)
        if value<root.value:
            if root.left is None:
                root.left = node
            else:
                insertAVL(root.left, value)
        elif value>=root.value:
            if root.right is None:
                root.right = node
            else:
                insertAVL(root.right, value)
    # check for AVL rules
    root = checkBalance(root)
    return root
```

```python
root = insertAVL(root, 54)
```

```python
lolviz.treeviz(root)
```
